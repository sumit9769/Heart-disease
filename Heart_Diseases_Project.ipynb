# Taking Care of Duplicate Values

```python
# Here we can check duplicate values if it is there then we have to drop them

df_dup=df.duplicated().any()

# let me create one variable data underscore loop is equal to data dot,we have to use dupicated 
# method of pandas data frame dot any let me.
```

```python
df_dup

# here output is true means our data set contains some diplicate values 
```

```python
# here we drop duplictes values 

df=df.drop_duplicates()  # and ,modify our existing dataframe
```

```python
df_dup

# we can se here output is false, now all the duplicated values are removed .
#we can see over here because output is false previously it was true .
#so this way we can check for duplicate values and this way we can drop them. 
```

```python
# first we seprate categorical columns and numerical columns means columns with categorical values and numerical values
# 1st we create two empty list .

cat_value=[]   #this for categorical columns 
con_value=[]   #this is for numerical columns

# you can give any name so let me write 4 columns name

for column in df.columns:
    if df[column].nunique() <=10: #here we also use nunique method of dataframe
        cat_value.append(column)
    else:
        con_value.append(column)
```

```python
cat_value  # if less than 10 than we consider as categorical columns which contain categorical values
```

```python
con_value # higher than 10 than we consider as numerical columns which contain numerical values
```

# Encoding of categorical Data

```python
cat_value
```

```python
df['cp'].unique()
```

```python
cat_value.remove('sex')
cat_value.remove('target')
df=pd.get_dummies(df,columns=cat_value,drop_first=True)
```

# SVC (SUPPORT VECTOR CLASSIFIER)

```python
from sklearn.svm import SVC
```

```python
svm=SVC()
svm.fit(X_train,Y_train)
ypred=svm.predict(X_test)

train=svm.score(X_train,Y_train)
test=svm.score(X_test,Y_test)

#print(f"Training accuracy : {train}\nTesting accuracy : {test}")
print(train)
print(test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

ac=accuracy_score(Y_test,ypred)
cm=confusion_matrix(Y_test,ypred)
cr=classification_report(Y_test,ypred)

print(f"Accuracy : {ac}\n\nConfusion Matrix :\n {cm}\n\nClassification Report :\n{cr}")

# as you can see over here our support vector classifier is around 80 percent on this particular dataset
```

# KNeigbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
# now lets train the model 
knn.fit(X_train,Y_train)
```

```python
# SO now lets perform prediction....

# So let me create one variable ...y_pred2..

y_pred2=knn.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred2)

# as you can see over here our KNeighbors classifier is around 74 percent accurate on this particular dataset 
```

```python
# Now here we will check for which value this k neighbors classifier can provide better accuracy
# 1st create 1 empty list

score=[]

for k in range (1,40):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,Y_train)
    y_pred=knn.predict(X_test)       
    score.append(accuracy_score(Y_test,y_pred))# here we store accuracy_score in score list
    
    
```

```python
# now here we can check score list

score

# we can see here accuracy score for different values of k.

# here k value 1 and  k value 2 .

# so this model can provide better accuracy if we k value = 2
```

```python
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,Y_train)
y_pred=knn.predict(X_test)       
accuracy_score(Y_test,y_pred)

# here we pass vale 2 in n_neighbors for better accuracy

# we can see that our accuracy score is increased by 80 percent previously it was 74 percent 

# So we can find better value for n_neighbors parameter of this k neigbors classifier  
```
